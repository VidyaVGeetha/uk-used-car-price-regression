```python
#Project Name - UK Used Car Price Prediction Using-  Regression.

#Regression - A regression plot helps you see how one variable changes with another,
# and whether the change looks linear, curved, or random.

#Step 1: Install required libraries.
!pip install pandas    #Data loading, cleaning.
!pip install numpy     #Numerical Calculations.
!pip install matplotlib # Data Visualization.
!pip install seaborn    #Beautiful Data Visualizations.

#Step 2: Import Libraries.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

 
```

```python
#Step 3: Read the dataset into a pandas DataFrame.
df_car=pd.read_csv("UK_Used_Car_Prices.csv")
print('Data read into a pandas dataframe!')
```

```python
#3.1: Preview the First Five Rows of the Dataset
df_car.head()
```

```python
#Step 4: Explore the Data (EDA Basics).
#4.1: Check for missing values in each column.
df_car.isnull().sum()
```

```python

# 4.2: View dataset structure and data types.
df_car.info()
```

```python
# 4.3: Check for Duplicates.
df_car.duplicated().sum()
```

```python
# 4.4: Explore categories(Check Unique Values in Categorical Columns)
print("Unique Car Makes:", df_car['Make'].unique())
print("Unique Models:" , df_car['Model'].unique())
print("Unique Fuel Types:",df_car['Fuel_Type'].unique())
```

```python
# 4.5: Statistical Summary of Key Numerical Features.
df_car.describe()
```

```python
#Step 5: Mileage vs Price  Regression Plot

#5.1: Create a Figure  (Create a Canvas for the Plot)
plt.figure(figsize=(10,6))

#5.2: Create a Regression Plot (Plot the Relationship Between Mileage and Price)
sns.regplot(x='Mileage',
            y='Price', 
            data=df_car,
            scatter_kws={'alpha':0.5},
            line_kws={'color':'red'}
           )
#5.3: Add a Title
plt.title("Regression Plot: Mileage vs Price")


#5.4: Show the plot
plt.show()

# Note: This regression plot shows a clear negative relationship between mileage and price.
# As mileage increases, the price of a used car decreases.
```

```python
# 5.5: Beautiful Regression Plot.
plt.figure(figsize=(12,7))

# Use a clean style
sns.set_style("whitegrid")

#Create a Regression Plot.
sns.regplot(
    x='Mileage', 
    y='Price', 
    data=df_car,
    scatter_kws={'alpha':0.6, 'color':'#1f77b4', 's':50},   # Customize the Appearance of the Scatter Plot Points(nice dots)
    line_kws={'color':'#d62728', 'linewidth':3} )            # Thicker Red line.
    
# Add a stylish title
plt.title("Mileage vs Price: Regression Analysis of UK Used Cars",
          fontsize=20, 
          color="Green", 
          weight='bold'
         )

#  Axis labels with styling.
plt.xlabel("Mileage (in km)", fontsize=18, color="Red", weight='bold')  
plt.ylabel("Price ()", fontsize=18, color="Red", weight='bold')

# Improve tick size
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Add tighter layout
plt.tight_layout()


plt.show()

#__________________________________________________

#The relationship between Mileage and Price SHOULD be negative.
#Therefore,  regression line must slope downward.
    
```

```python
# Step 6: Year vs Price  Regression Plot.

# 6.1: Create a Figure  (Create a Canvas for the Plot)
plt.figure(figsize=(10,7))

# 6.2: Set the Plot Style to Dark Grid
sns.set_style("darkgrid")

# 6.3: Create Regression Plot
sns.regplot( x='Year' ,
            y='Price' ,
            data=df_car,
            scatter_kws={'alpha':0.6, 'color':'#17becf', 's':55},   # Customize the Appearance of the Scatter Plot Points(nice dots)
            line_kws={'color':'#ff7f0e', 'linewidth':3} )           # Bold professional color.
            
    
# 6.4:  Add title
plt.title("Regression Plot: Year vs Price",
         fontsize=24, 
          color="Green", 
          weight='bold')

# 6.5: Axis labels with styling.
plt.xlabel("Manufacturing Year", fontsize=18,color="Blue")
plt.ylabel("Price (GBP )", fontsize=18, color="Blue")

# 6.6:  Improve tick size.
plt.xticks(fontsize=12)
#plt.xticks(df_car['Year'].unique(), rotation=45)
plt.yticks(fontsize=12)


# 6.7: Add tighter layout.
plt.tight_layout()

# 6.8:Show plot.
plt.show()
#_______________________________________________________
#Note: The regression line slopes upward 
#Newer cars (higher year) tend to have higher prices.

```

```python
# **** Display Year as Integers on the X-Axis ***

# Step 6: Year vs Price  Regression Plot.

# 6.1: Create a Figure  (Create a Canvas for the Plot)
plt.figure(figsize=(10,7))

# 6.2: Set the Plot Style to Dark Grid
sns.set_style("darkgrid")

# 6.3: Create Regression Plot
sns.regplot( x='Year' ,
            y='Price' ,
            data=df_car,
            scatter_kws={'alpha':0.6, 'color':'#17becf', 's':55},   # Customize the Appearance of the Scatter Plot Points(nice dots)
            line_kws={'color':'#ff7f0e', 'linewidth':3} )           # Bold professional color.
            
    
# 6.4:  Add title
plt.title("Regression Plot: Year vs Price",
         fontsize=24, 
          color="Green", 
          weight='bold')

# 6.5: Axis labels with styling.
plt.xlabel("Manufacturing Year", fontsize=18,color="Blue", weight='bold')
plt.ylabel("Price (GBP )", fontsize=18, color="Blue", weight='bold')

# 6.6:  Improve tick size.
plt.xticks(df_car['Year'].unique(), rotation=45)
plt.yticks(fontsize=12)


# 6.7: Add tighter layout.
plt.tight_layout()

# 6.8:Show plot.
plt.show()
#_______________________________________________________
#Note: The regression line slopes upward 
#Newer cars (higher year) tend to have higher prices.


```

```python
# Step 7: Engine Size vs Price  Regression Plot.

# 7.1: Create a Figure  (Create a Canvas for the Plot)
plt.figure(figsize=(12,7))

# 7.2: Setting the Plot Style to ticks for a Clean, Minimal Look.
sns.set_style("ticks")

# 7.3: Create Regression Plot.
sns.regplot(x='Engine_Size_L',
           y='Price',
           data=df_car,
           scatter_kws={'alpha':0.6, 'color':'#9467bd', 's':55},
            line_kws={'color':'#ffa600', 'linewidth':3}
           )

# 7.4: Add Title
plt.title("Engine Size vs Used Car Price  Linear Regression Analysis",
           
           fontsize=24, 
           color="Red", 
           weight='bold'
         
         )

# 7.5: Axis labels with styling.
plt.xlabel("Engine Size (Litres)", fontsize=16, color="Green", weight='bold')
plt.ylabel("Price (GBP )", fontsize=16, color="Green", weight='bold')

# 7.6: Improve tick size.
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

# 7.7: Show plot
plt.show()
# _____________________________________________________

# Note: Cars with larger engine sizes tend to have higher prices, but the relationship is weaker than Year vs Price.

```
